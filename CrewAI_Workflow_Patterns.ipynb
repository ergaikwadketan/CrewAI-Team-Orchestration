### **What is CrewAI?**
Most people use LLMs (like ChatGPT or Gemini) as a **Chatbot**: You ask one question, it gives one answer.
**CrewAI** is different. It is a framework designed to build **AI Teams**.

Imagine you are running a software company. You wouldn't hire one person to be the CEO, Coder, Designer, and HR all at once. You would hire specialists. CrewAI allows you to do exactly that with AI.

> **Core Philosophy:** "Don't ask one smart AI to do everything. Ask a team of specialized AI Agents to work together."

---

### **ğŸ§± The 4 Building Blocks of CrewAI**

To build a team, you need four things. Think of this like a movie production crew:

#### **1. Agents (The "Who")**
Agents are your employees. They are not just generic AI; they are role-players.
* **Role:** What is their job title? (e.g., "Senior Researcher")
* **Goal:** What are they trying to achieve? (e.g., "Uncover hidden facts")
* **Backstory:** This gives them personality and context! (e.g., "You are a veteran journalist who hates fake news.")
* *Analogy:* The "Scriptwriter" or the "Director" on a movie set.

#### **2. Tasks (The "What")**
Agents don't guess what to do; you give them specific assignments.
* **Description:** Clear instructions on what needs to be done.
* **Expected Output:** What does "done" look like? (e.g., "A 4-paragraph blog post" or "A JSON file").
* *Analogy:* "Write the opening scene" or "Scout locations for filming."

#### **3. Process (The "Workflow")**
How does the team collaborate?
* **Sequential:** Like a relay race. Agent A finishes, passes the baton to Agent B. (Most common).
* **Hierarchical:** Like a corporate office. A "Manager" AI assigns tasks to workers, reviews their work, and asks for changes.
* *Analogy:* Does the Writer finish the script before the Actor starts (Sequential)? Or does the Director shout orders to everyone at once (Hierarchical)?

#### **4. Tools (The "How")**
Agents can be smart, but they are blind without tools. Tools give them "superpowers" to interact with the real world.
* **Examples:** Google Search, Calculator, YouTube Scraper, File Reader.
* *Analogy:* Giving the Cameraman a camera, or the Editor a computer.

```python
import os
from crewai import LLM, Agent, Task, Crew, Process
```

```python
# ğŸ”‘ CONFIGURATION
# Replace with your actual Google Gemini API Key
os.environ["GOOGLE_API_KEY"] = "Your_API_KEY"

# Initialize the Brain (LLM)
# We use temperature=0.2 for a balance of creativity and factual accuracy
my_llm = LLM(model="gemini/gemini-2.5-flash", temperature=0.2)

print("âœ… CrewAI Installed and LLM Ready.")
```

```python
# ==========================================
# SCENARIO 1: SEQUENTIAL PROCESS
# Context: Research a festival -> Write a post
# ==========================================

# --- 1. Define Agents (The Workers) ---
researcher = Agent(
    role='Cultural Researcher',
    goal="Find detailed and unique facts about {festival}",
    backstory="You are an expert in Indian culture and festivals. You love finding obscure facts.",
    llm=my_llm,
    verbose=True
)

writer = Agent(
    role='Social Media Writer',
    goal="Write a viral LinkedIn post using the researcher's facts",
    backstory="You are a top-tier ghostwriter for influencers. You write engaging, fun content.",
    llm=my_llm,
    verbose=True
)

# --- 2. Define Tasks (The To-Do List) ---
# Task 1: Get the data
task_research = Task(
    description='Find 3 unique and interesting facts about the {festival}.',
    expected_output="A bulleted list of 3 facts.",
    agent=researcher
)

# Task 2: Use the data (Context ensures Task 1's output is passed here)
task_post = Task(
    description='Write a LinkedIn post using the facts found. Add relevant hashtags.',
    expected_output="A professional LinkedIn post.",
    agent=writer,
    context=[task_research] # <--- Crucial: This passes the research to the writer
)

# --- 3. Create the Crew (The Team) ---
sequential_crew = Crew(
    agents=[researcher, writer],
    tasks=[task_research, task_post],
    process=Process.sequential, # Flow: Research -> Write
    verbose=True
)

# --- 4. Execution ---
print("\nğŸ Kicking off Sequential Crew...")
result = sequential_crew.kickoff(inputs={'festival': 'Onam'})
print(f"\nğŸ“¢ Final Result:\n{result}")
```

```python
# ==========================================
# SCENARIO 2: HIERARCHICAL PROCESS (MANAGER MODE)
# Context: Research -> Delegate to multiple specialized writers
# ==========================================

# --- 1. Define Specialized Agents ---
# (We reuse the researcher from above)

linkedin_writer = Agent(
    role='LinkedIn Specialist',
    goal="Write professional articles for LinkedIn",
    backstory="You write structured, professional content for business audiences.",
    llm=my_llm,
    verbose=True
)

twitter_writer = Agent(
    role='Twitter Specialist',
    goal="Write punchy tweets under 280 characters",
    backstory="You are a master of brevity and emojis. You write viral tweets.",
    llm=my_llm,
    verbose=True
)

# --- 2. Define Tasks ---
task_research_h = Task(
    description='Find 3 unique facts about the {festival}.',
    expected_output="List of facts.",
    agent=researcher
)

task_linkedin = Task(
    description='Write a professional LinkedIn post using the research.',
    expected_output="A structured LinkedIn post.",
    agent=linkedin_writer,
    context=[task_research_h]
)

task_twitter = Task(
    description='Write a fun Tweet using the research. Must be under 280 chars.',
    expected_output="A short tweet with emojis.",
    agent=twitter_writer,
    context=[task_research_h]
)

# --- 3. Create the Managerial Crew ---
# Note: In Hierarchical process, we MUST define a 'manager_llm'
manager_crew = Crew(
    agents=[researcher, linkedin_writer, twitter_writer],
    tasks=[task_research_h, task_linkedin, task_twitter],
    process=Process.hierarchical, # Flow: Manager decides the order
    manager_llm=my_llm,           # The "Brain" of the manager
    verbose=True
)

# --- 4. Execution ---
print("\nğŸ Kicking off Hierarchical Crew...")
result = manager_crew.kickoff(inputs={'festival': 'Diwali'})
print(f"\nğŸ“¢ Final Result:\n{result}")
```

```python
# ==========================================
# SCENARIO 3: DYNAMIC INPUTS (REUSABILITY)
# Context: One Agent behaving differently based on inputs
# ==========================================

# --- 1. Flexible Agent ---
# Notice we use {social_platform} in the backstory/goal
flexible_writer = Agent(
    role='Social Media Manager',
    goal="Write viral content specifically for {social_platform}",
    backstory="You are an expert at adapting your writing style for {social_platform}.",
    llm=my_llm,
    verbose=True
)

# --- 2. Flexible Task ---
task_flexible_post = Task(
    description='Write a post about {festival} specifically tailored for {social_platform}.',
    expected_output="A post optimized for the chosen platform.",
    agent=flexible_writer
)

# --- 3. Create Crew ---
dynamic_crew = Crew(
    agents=[flexible_writer],
    tasks=[task_flexible_post],
    verbose=True
)

# --- 4. Execution Examples ---

# Example A: Twitter
print("\nğŸ Running for Twitter...")
tweet_result = dynamic_crew.kickoff(inputs={'festival': 'Holi', 'social_platform': 'Twitter'})
print(f"ğŸ¦ Tweet:\n{tweet_result}")

# Example B: LinkedIn
print("\nğŸ Running for LinkedIn...")
linkedin_result = dynamic_crew.kickoff(inputs={'festival': 'Holi', 'social_platform': 'LinkedIn'})
print(f"ğŸ’¼ LinkedIn Post:\n{linkedin_result}")
```
